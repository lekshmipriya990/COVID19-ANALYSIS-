```python
#import the modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print("Modules are imported")
```

```python
#import the dataset for covid cases
corona_dataset_csv = pd.read_csv(r"Datasets\covid19_deaths_dataset.csv")
corona_dataset_csv.head(10)
```

```python
#size of the dataset
corona_dataset_csv.shape
```

```python
corona_dataset_csv.head()
```

```python
#remove useless columns
corona_dataset_csv.drop(["Lat","Long"], axis=1, inplace=True)
corona_dataset_csv.head()
```

```python
#aggregate by country
corona_data_aggregated = corona_dataset_csv.groupby("Country/Region").sum()
```

```python
corona_data_aggregated.head()
```

```python
corona_data_aggregated.loc["India"]
```

```python
#plotting the graph
corona_data_aggregated.loc["China"].plot()
corona_data_aggregated.loc["India"].plot()
corona_data_aggregated.loc["Italy"].plot()
plt.legend()
```

```python
corona_data_aggregated.loc["India"].plot()
```

```python
#plotting the required date
corona_data_aggregated.loc["India"][50:53].plot()
```

```python
#derivative 
corona_data_aggregated.loc["India"].diff().plot()
```

```python
#maximum death in India in one day
corona_data_aggregated.loc["India"].diff().max()
```

```python
#Finding the max infection rate for all countries
countries = list(corona_data_aggregated.index)
max_death_rate= []
for c in countries :
    max_death_rate.append(corona_data_aggregated.loc[c].diff().max())
corona_data_aggregated["max_death"] = max_death_rate
```

```python
corona_data_aggregated.head()
```

```python
corona_data = pd.DataFrame(corona_data_aggregated["max_death"])
```

```python
corona_data.head()
```

```python
#importing happiness report dataset
happiness_report_csv = pd.read_csv(r"Datasets\worldwide_happiness_report.csv")
happiness_report_csv.head()
```

```python
#removing the useless columns
useless_cols = [ "Overall rank", "Score", "Generosity", "Perceptions of corruption"]
```

```python
happiness_report_csv.drop(useless_cols,axis=1,inplace=True)
happiness_report_csv.head()
```

```python
#making the country/region as index
happiness_report_csv.set_index( "Country or region",inplace=True)
happiness_report_csv.head()
```

```python
corona_data.head()
corona_data.shape
```

```python
#joining both datasets
data = corona_data.join(happiness_report_csv,how= "inner")
data.head()
```
